```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=180)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('BNB-USD',
                      start=start_date,
                      end=end_date,
                      progress=False)
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
```

```python
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"],
                                        high=data["High"],
                                        low=data["Low"],
                                        close=data["Close"])])
figure.update_layout(title = "BNB Price Analysis",
                     xaxis_rangeslider_visible=False)
figure.show()
```

```python
correlation = data.corr()
print(correlation["Close"].sort_values(ascending=False))
```

```python
pip install autots
```

```python
from autots import AutoTS

model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```

```python
# Step 4: Membuat prediksi dalam format candlestick
# Prediksi hanya memberikan 'Close', jadi kita membuat perkiraan untuk 'Open', 'High', dan 'Low'
forecast["Open"] = forecast["Close"].shift(1)
forecast["Open"].iloc[0] = data["Close"].iloc[-1]  # Open pertama prediksi sama dengan Close terakhir
forecast["High"] = forecast[["Open", "Close"]].max(axis=1) * 1.02  # Tingkatkan 2% sebagai asumsi High
forecast["Low"] = forecast[["Open", "Close"]].min(axis=1) * 0.98   # Kurangi 2% sebagai asumsi Low
```

```python
# Menambahkan tanggal ke prediksi
forecast['Date'] = pd.date_range(start=data['Date'].iloc[-1] + timedelta(days=1), periods=30, freq='D')
```

```python
# Step 6: Visualisasi candlestick untuk data historis dan prediksi
figure.add_trace(go.Candlestick(x=forecast["Date"],
                                open=forecast["Open"],
                                high=forecast["High"],
                                low=forecast["Low"],
                                close=forecast["Close"],
                                name='Prediction',
                                increasing_line_color='cyan',
                                decreasing_line_color='gray'))

figure.update_layout(title="BNB-USD Historical and Predicted Prices",
                     xaxis_title="Date",
                     yaxis_title="Price (USD)",
                     xaxis_rangeslider_visible=False)

# Tampilkan chart
figure.show()
```
